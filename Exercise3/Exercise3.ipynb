```python
# Correlation Time

def corr(acf, time_values):
    '''
    Calculate the correlation time based on the autocorrelation function.

    Parameters:
    acf (array-like): Autocorrelation function values.
    time_values (array-like): Corresponding time values.

    Returns:
    float: The correlation time.
    '''
    # Find the first index where the autocorrelation decays to e^-1
    e_minus_1 = np.exp(-1)
    idx = np.argmax(acf < e_minus_1)

    # Interpolate to find the exact correlation time
    if idx == 0:
        return 0.0
    else:
        slope = (acf[idx] - acf[idx - 1]) / (time_values[idx] - time_values[idx - 1])
        intercept = acf[idx] - slope * time_values[idx]
        correlation_time = (np.log(e_minus_1) - intercept) / slope
        return correlation_time

time = np.linspace(0, 1000, trj.n_frames)
lags = range(10)

phi8_acf = autocorr1(dihedral_dict['phi'][8], lags)
corr_phi8 = corr(phi8_acf, time)
print("Correlation time for phi8:", corr_phi8)
```

```python
# 4) Calculate the ciruclar correlation coefficients 
# between phi and psi angles and reproduce the correlation map

def circular_mean(angle):
    return np.arctan2(np.sum(np.sin(angle)),np.sum(np.cos(angle)))

def circular_corr(angle1,angle2):
    cmean1=circular_mean(angle1)
    cmean2=circular_mean(angle2)
    numerator=(np.sum(np.sin(angle1-cmean1)*np.sin(angle2-cmean2)))
    denom=((np.sum(np.sin(angle1-cmean1)**2)*np.sum(np.sin(angle2-cmean2)**2))**.5)
    ccor=numerator/denom
    return ccor

# Make correlation contact map
correlation_maps = []
for i in range(43, 76):
    cor_map = []
    for j in range(43, 76):
        phi=dihedral_dict['phi'][i]
        psi=dihedral_dict['psi'][j]
        ccor=circular_corr(psi,phi)
        cor_map.append(ccor)
    correlation_maps.append(cor_map)
final_cor_map = np.asarray(correlation_maps).astype(float)
np.shape(final_cor_map)

sequence=[]
for residue in trj.topology.residues:
    sequence.append(residue)
len(sequence)
seq_trunc=sequence[43:76]
seq_trunc_length = len(seq_trunc)
#Lets Zero Out some of the Labels 
n=2
for i in range(42, min(75, seq_trunc_length)):
    if (i % n !=0):
        seq_trunc[i]=''

mask = np.zeros_like(final_cor_map)
zeros=np.where((final_cor_map<0.10) & (final_cor_map>-0.10))
mask[zeros] = True
ax = sns.heatmap(final_cor_map,mask=mask, vmin=-0.5, vmax=0.5,cmap='seismic')
ax.invert_yaxis()
ax.grid(which='both', alpha=0.5)
ax.set_xticks(range(0,33))
ax.set_yticks(range(0,33))
ax.set_xticklabels(seq_trunc,rotation=45)
ax.set_yticklabels(seq_trunc,rotation=45)
plt.show()
```
